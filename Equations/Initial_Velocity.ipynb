# Solve for Initial Velocity ($v_0$):

# $$v_0 = v - at \quad \mid \quad v_0 = \frac{\Delta x}{t} - \frac{1}{2}at \quad \mid \quad v_0 = \sqrt{v^2 - 2a\Delta x}$$


# 1) 

$$v_0 = v - at$$

Where:

$v$: This is your final target velocity at the very end of the timeline. It represents the total accumulated speed you are currently holding, which includes both your original starting speed and any extra speed gained along the way.

$- at$: This is the time-subtraction anchor. Multiplying acceleration ($a$) by time ($t$) calculates the exact amount of speed your engine manufactured during the trip. Subtraction ($\mathbf{-}$) acts as a rewind button on the clock, actively stripping that manufactured speed away from your final velocity to isolate what your baseline speed was before you ever stepped on the gas.

---

# 2) 

$$v_0 = \frac{\Delta x}{t} - \frac{1}{2}at$$

Where:

$\frac{\Delta x}{t}$: This calculates your crude average velocity across the entire trip. It simply takes the total spatial gap crossed ($\Delta x$) and distributes it uniformly over the total time elapsed ($t$), blindly assuming you traveled at one single, steady speed the whole time.

$- \frac{1}{2}at$: This is the ramp-up correction factor. Because you were actually accelerating, your speed was constantly rising, meaning your crude average velocity is falsely inflated. Subtraction ($\mathbf{-}$) strips away exactly half ($\frac{1}{2}$) of the total speed gained from acceleration ($at$), pulling that inflated average down to reveal your true, raw velocity at the exact moment the timeline began.

---

# 3) 

$$v_0 = \sqrt{v^2 - 2a\Delta x}$$

Where:

$v^2$: This is your final, maximum speed capacity squared. It represents the total physical ceiling of raw energy capacity you possess at the very end of your spatial journey.

$- 2a\Delta x$: This is the spatial-subtraction engine. Multiplying your acceleration rate ($a$) by the spatial gap crossed ($\Delta x$) calculates the total intensity of speed capacity manufactured over that distance, while the $2$ acts as the structural scaling constant. Subtraction ($\mathbf{-}$) strips this entire block of distance-gained capacity away from your final ceiling.

$\sqrt{\dots}$ [The Square Root]: The radical acts as the dimensional converter. Because the subtraction step happens in "squared speed" units, the square root steps in at the very end to crush that non-linear energy capacity back down into a standard, live velocity metric (meters per second).
