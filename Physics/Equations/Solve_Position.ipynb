# Definition ($x_f$):

# $$x_f = x_i + vt$$

$x_f$ = Final Position (Where you end up)  

$x_i$ = Initial Posiiton (Where you started)  

$v * t, vt$ = Constant Velocity (How fast you are moving (magnitude) and in what direction = Vector) * Time Elapsed (How long you have been moving - start to end)  

Eg: Positive Velocity (Forward Motion)

- Initial position $(x_i) = 5m$
- Constant velocity $(v) = 3m/s$
- Time elapsed $(t) = 4s$

Solve:
- $x_f = x_i + \left(v_{\left[\frac{\text{m}}{\text{s}}\right]} \times t_{[\text{s}]}\right) \quad \rightarrow \quad x_f = x_i + \text{Distance}_{[\text{m}]} \quad \rightarrow \quad x_f = 17\text{m}$

Where: 
- $x_f = 5 + (3 * 4)$
- $x_f = 17m$

# Solve Postion ($x_f$):

# 1) $$x_f = x_i + v_0t + \frac{1}{2}at^2$$  

#### You use this equation when Time ($t$) is actively involved in the problem (either given to you or what you are trying to find), but you have absolutely no idea what the Final Velocity ($v$) is. "An object drops from a building. Where is it after 3 seconds?", "How long does it take a braking car to slide 50 meters?", notice that the variable for final velocity ($v$) is completely missing from this equation. It doesn't care how fast the object is moving at the end of the trip.  

Where:

$x_i$: Before anything moves, the equation needs to know where you are standing on the map when the stopwatch clicks. This is $x_i$ (Initial Position). If you start your experiment $10\text{ meters}$ down the track from the zero-marker, $x_i = 10\text{ m}$. This is your static spatial launchpad. Everything else we calculate will be stacked on top of this number.

$v_0 * t$: Now, the clock starts ticking. You are traveling at a perfectly constant initial velocity ($v_0$). Why are $v_0$ and $t$ multiplied together? Because multiplication means accumulation over time.

$\frac{1}{2}at^2$: The $at^2$ part calculates the maximum possible distance if you were going fast the whole time. The $\frac{1}{2}$ steps in to cut that number in half, accurately reflecting your gradual ramp-up.

# 2) $$x_f = x_i + \frac{v^2 - v_0^2}{2a}$$

#### You use this equation when Time ($t$) is completely missing from the conversation. You don't know the time, and the question doesn't ask you to find the time. This is your ultimate shortcut tool. If an exam question gives you speeds and distances but keeps the clock hidden, this equation allows you to completely bypass calculating how many seconds passed. "A car traveling at 30 m/s slams on the brakes and comes to a stop. What distance does it slide?", "A diver jumps from a 10-meter platform (Question 8 on your exam printout!). How fast are they moving when they hit the water?".

Where:  

$x_i$: Before anything moves, the equation needs to know where you are standing on the map when the stopwatch clicks. This is $x_i$ (Initial Position). If you start your experiment $10\text{ meters}$ down the track from the zero-marker, $x_i = 10\text{ m}$. This is your static spatial launchpad. Everything else we calculate will be stacked on top of this number.  

$v^2 - v_0^2$: This tracks your physical change in raw energy capacity. Because acceleration forces your velocity to scale up non-linearly, squaring the speeds ($v^2$ and $v_0^2$) is the only way to measure the true intensity of the engine's work. Subtraction actively strips away your starting speed threshold ($v_0^2$), isolating the pure, raw amount of speed capacity manufactured during the sprint.  

$\frac{1}{2a}$: This is the spatial divider. Since this equation completely ignores the clock, it cannot use seconds to distribute that speed capacity. Instead, it divides by your constant acceleration rate ($a$) to transform that abstract speed change directly into a real physical coordinate. The $2$ on the bottom acts as the average-marker, compensating for the fact that you weren't traveling at top speed the whole time, perfectly converting your engine's effort into a raw spatial gap.
